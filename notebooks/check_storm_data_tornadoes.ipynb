```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
duration = (30-17)*60 
date = '20180502' #'20210504', '20180502'
initial_time = f'{date}1700'
```

```python
dtype = {'VALID': np.int64, 'LAT':np.float64, 'LON':np.float64, 
        'MAG':np.float64, 'EVENT_TYPE':object, 'TOR_F_SCALE' :object}

df = pd.read_csv('/work/mflora/LSRS/STORM_DATA_2017-2020.csv', dtype=dtype, na_values = 'None')
df['date'] = pd.to_datetime(df.VALID.astype(str), format='%Y%m%d%H%M')
```
