```python
import pandas as pd 
import numpy as np
import itertools
```

```python
times = ['first_hour', 'second_hour']
modes = ['training', 'testing']
target_cols = ['tornado', 'severe_hail', 'severe_wind']

for time, mode, target_col in itertools.product(times, modes, target_cols):
    df = pd.read_feather(
        f'/work/mflora/ML_DATA/DATA/original_{time}_{mode}_matched_to_{target_col}_0km_data.feather')
    
    dates = np.unique(df['Run Date'])
    df = pd.DataFrame({'Dates' : dates})
    
    df.to_feather(f'/work/mflora/ML_DATA/DATA/original_dates/{mode}_{time}_{target_col}_dates.feather')
```
